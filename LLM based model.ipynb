```python
import pandas as pd
import streamlit as st
import spacy
import re

# Ensure that the 'spacy' library is installed and the 'en_core_web_sm' model is downloaded.
# Run 'pip install spacy' and 'python -m spacy download en_core_web_sm' if not already installed.
nlp = spacy.load("en_core_web_sm")

# Define a function to extract structured data
def extract_data_from_notes(notes):
    structured_data = []

    for note in notes:
        doc = nlp(note)

        # Extract age
        age_match = re.search(r"(\d+)-year-old", note)
        age = age_match.group(1) if age_match else None

        # Extract gestational age using regex
        gestation_match = re.search(r"\b(\d+) weeks? gestation\b", note)
        gestational_age = gestation_match.group(1) if gestation_match else None

        # Extract symptoms
        symptoms = re.findall(r"\b(reports|complains of) (.+?)(\.|,| and)", note)
        symptoms_list = [sym[1] for sym in symptoms]

        # Extract medical history
        history_match = re.search(r"No history of (.+?)\.", note)
        history = history_match.group(1) if history_match else "None"

        # Extract lab results using keyword matching
        lab_results = re.findall(r"(hemoglobin|glucose tolerance).+?\b(normal|high|low)\b", note)
        lab_results_dict = {lab[0]: lab[1] for lab in lab_results}

        # Extract medications
        medication_match = re.search(r"Prescribed (.+?)\.", note)
        medications = medication_match.group(1) if medication_match else None

        # Append structured data
        structured_data.append({
            "Age": age,
            "Gestational Age": gestational_age,
            "Symptoms": ", ".join(symptoms_list),
            "History": history,
            "Lab Results": lab_results_dict,
            "Medications": medications,
        })

    return pd.DataFrame(structured_data)

# Add predictions for fetal health and gestational term
def add_predictions(data):
    predictions = []
    for _, row in data.iterrows():
        prediction = {
            "Fetal Health": "Normal" if row["Lab Results"].get("hemoglobin", "normal") == "normal" else "At Risk",
            "Gestational Term": "Full Term" if row["Gestational Age"] and int(row["Gestational Age"]) >= 37 else "Preterm"
        }
        predictions.append(prediction)
    predictions_df = pd.DataFrame(predictions)
    return pd.concat([data, predictions_df], axis=1)

# Streamlit app
def main():
    st.title("EMR Analysis and Prediction App")

    # File upload
    uploaded_file = st.file_uploader("Upload a file (text or batch file in CSV format)", type=["txt", "csv"])

    if uploaded_file is not None:
        # Handle single text file
        if uploaded_file.name.endswith(".txt"):
            content = uploaded_file.read().decode("utf-8")
            notes = [content]
        elif uploaded_file.name.endswith(".csv"):
            df = pd.read_csv(uploaded_file)
            notes = df["Notes"].tolist()  # Assuming a column named 'Notes' contains unstructured EMR data

        # Process and predict
        structured_data = extract_data_from_notes(notes)
        results = add_predictions(structured_data)

        # Display results
        st.subheader("Structured Data with Predictions")
        st.dataframe(results)

        # Download results
        csv = results.to_csv(index=False).encode('utf-8')
        st.download_button(
            label="Download Results as CSV",
            data=csv,
            file_name="emr_predictions.csv",
            mime="text/csv",
        )

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
from docx import Document

uploaded_file_path = '/Users/rajeevkumar/Documents/Internship portfolio/Phase 3_LLM/EMR_Form.docx'
# Function to extract content from a Word document
def extract_emr_form_content(file_path):
    doc = Document(file_path)
    content = []
    for paragraph in doc.paragraphs:
        if paragraph.text.strip():  # Only add non-empty paragraphs
            content.append(paragraph.text.strip())
    return content

# Function to process extracted content into structured data
def process_emr_content(content):
    structured_data = []
    for line in content:
        # Example parsing logic (adjust based on your form structure)
        if "Age:" in line:
            age = int(line.split(":")[1].strip())
        elif "Gestational Age:" in line:
            gestational_age = int(line.split(":")[1].strip())
        elif "Hemoglobin:" in line:
            hemoglobin = line.split(":")[1].strip()
            structured_data.append({
                "Age": age,
                "Gestational Age": gestational_age,
                "Hemoglobin": hemoglobin
            })
    return pd.DataFrame(structured_data)

# Function to add predictions and recommendations
def add_predictions(data):
    predictions = []
    recommendations = []
    for _, row in data.iterrows():
        fetal_health = "Normal" if row["Hemoglobin"] == "normal" else "At Risk"
        gestational_term = "Full Term" if row["Gestational Age"] >= 37 else "Preterm"
        
        # Generate recommendations
        recommendation = []
        if fetal_health == "At Risk":
            recommendation.append("Monitor fetal health closely and consult a specialist.")
        if gestational_term == "Preterm":
            recommendation.append("Schedule regular prenatal visits to monitor progression.")
        if fetal_health == "Normal" and gestational_term == "Full Term":
            recommendation.append("Maintain a healthy diet and follow standard prenatal care.")
        
        predictions.append({
            "Fetal Health": fetal_health,
            "Gestational Term": gestational_term
        })
        recommendations.append(" ".join(recommendation))
    
    predictions_df = pd.DataFrame(predictions)
    recommendations_df = pd.DataFrame({"Recommendations": recommendations})
    return pd.concat([data, predictions_df, recommendations_df], axis=1)

# Main function to process a `.docx` file and generate output
def main(file_path):
    content = extract_emr_form_content(file_path)
    structured_data = process_emr_content(content)
    results = add_predictions(structured_data)
    print(results)

    # Save the results to a CSV file
    results.to_csv("emr_predictions.csv", index=False)
    print("Results saved to emr_predictions.csv")

# Replace 'EMR_Form.docx' with the path to your file
file_path = "EMR_Form.docx"
main(file_path)
```
